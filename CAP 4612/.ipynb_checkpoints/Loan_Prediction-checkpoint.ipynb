### Encoding

```python
df = loan_data.copy()
df.head()
```

```python
df['Gender'].replace('Male', 1, inplace=True)
df['Gender'].replace('Female', 0, inplace=True)
df.head()
```

```python
df.Married = df.Married.map({'Yes':1, 'No':0})
df.head()
```

```python
df['Married'].value_counts()
```

```python
df2 = loan_data.copy()
```

```python
df2.head()
```

```python
df2.info()
```

```python
#encoding

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder


## Again we have to temporarily encode variables
lb = LabelEncoder()

cat_cols = [col for col in df2.columns if df2[col].dtype == 'object']

for col in cat_cols:
    df2[col] = lb.fit_transform(df2[col])
```

```python
df2.head()
```

```python
df2.info()
```

### Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
y = df2.iloc[1:289, 11]
x = df2.iloc[1:289, 0:10]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 3)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)

lr_train_predictions = lr_model.predict(x_train)
lr_test_predictions = lr_model.predict(x_test)


print('Train Accurary Score = ', metrics.accuracy_score(lr_train_predictions, y_train))
print('Test Accurary Score = ', metrics.accuracy_score(lr_test_predictions, y_test))
```

```python
lr_model = LogisticRegression( max_iter = 10000)
lr_model.fit(x_train, y_train)

lr_train_predictions = lr_model.predict(x_train)
lr_test_predictions = lr_model.predict(x_test)


print('Train Accurary Score = ', metrics.accuracy_score(lr_train_predictions, y_train))
print('Test Accurary Score = ', metrics.accuracy_score(lr_test_predictions, y_test))
```
