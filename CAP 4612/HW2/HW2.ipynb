```python
# 10. Write the code that outputs all the percentage of missing data for the features listed in step 8.

def get_numerical_summary(df):
    total = df.shape[0]
    columns_with_missing_data = [col for col in df.columns if df[col].isnull().sum() > 0]
    missing_percent = {}

    for col in columns_with_missing_data:
        null_count = df[col].isnull().sum()
        per = (null_count/total) * 100
        missing_percent[col] = per
        print("{} : {} ({}%)".format(col, null_count, round(per, 3)))
    return missing_percent

missing_percent = get_numerical_summary(df)
```

```python
# 11. Write the code that drops features that have more than 22% of their data missing from the 
# cancer_data dataframe.  Create a new dataframe named  cancer_data_cleaned that has all the 
# features that have less than 22% of their data missing.  Show the percentage of missing data of 
# the cancer_data_cleaned dataframe.

def drop_feature(data_frame, threshold):
    
    for col, per in missing_percent.items():
        if per > threshold:
            data_frame.drop(col, axis = 1, inplace = True)
            
    return data_frame

cancer_data_cleaned = drop_feature(df.copy(), 20)

_ = get_numerical_summary(cancer_data_cleaned)
```

```python
# 12. Drop all rows with a threshold of 1, print out the information shown below on the 
# cancer_data_cleaned dataframe.

def drop_row(df, threshold):
    # Getting Missing count of each sample            
    for idx in range(df.shape[0]):
        df.loc[idx, 'missing_count'] = df.iloc[idx, :].isnull().sum()  

    
    df.drop(df[df['missing_count'] > threshold].index, axis = 0, inplace = True)
   
    return df

print("Samples Before Removal : {}".format(cancer_data_cleaned.shape[0]))
cancer_data_cleaned = drop_row(cancer_data_cleaned, 1)
print("Samples After Removal : {}".format(cancer_data_cleaned.shape[0]))
```

```python
# 16.  Write a function that automates the logistic regression with test size from  0.1 – 0.9. It 
# returns a dataframe named report_dataframe.


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

def automation_lr(x, y):
    report_df = pd.DataFrame(columns=['Test_Size',
                                      'Train_Accuracy_Score',
                                      'Test_Accuracy_Score',
                                      'Normalize_Score'])
    model = LogisticRegression(max_iter=10000)
    
    sizes = np.arange(0.1, 1.0, 0.1)
    
    for loc, size in enumerate(sizes):
        x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle = True, train_size = size)
        
        # fit the model -> build the training model
        model.fit(x_train, y_train)
        
        # getting the test and trainset info for predictions
        predictions_train = model.predict(x_train)
        predictions_test = model.predict(x_test)
        
        # getting the test and trainset info for accuracy
        accuracy_train = metrics.accuracy_score(predictions_train, y_train)
        accuracy_test = metrics.accuracy_score(predictions_test, y_test)
        
        # normalizing the score
        normalized_score = accuracy_test/accuracy_train
        
        report_df.loc[loc] = [size,
                             accuracy_train,
                             accuracy_test,
                             normalized_score]
        
    return report_df
```

```python
# 17. Call automation_lr( x, y) with x being features 2-31 and y being feature 1 of the 
# cancer_data_cleaned   dataframe. Look at step 14 to see where I am getting the feature numbers. 
# Then output the report_dataframe after the function call. 

x = cancer_data_cleaned.iloc[:,2:31]
y = cancer_data_cleaned.iloc[:,1]

report = automation_lr(x, y)
report
```

```python
# 18. Pick a test size that you think is the best then re-estimate the Logistic Regression with x 
# being features 2-31 and y being feature 1. Output the following:

logregression_model = LogisticRegression(max_iter=10000)
x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle = True, train_size = 0.8)

logregression_model.fit(x_train, y_train)

prediction_train = logregression_model.predict(x_train)
prediction_test = logregression_model.predict(x_test)

accuracy_train = metrics.accuracy_score(prediction_train, y_train)
accuracy_test = metrics.accuracy_score(prediction_test, y_test)

print(f"Train Accuracy Score = {accuracy_train}")
print(f"Test Accuracy Score = {accuracy_test}")
```
